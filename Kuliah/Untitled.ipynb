```python
import numpy as np

def iterate(a):
    n = 100
    w = np.zeros(n+1)
    x = np.zeros(n+1)
    y = np.zeros(n+1)
    v = np.zeros(n+1)
    y[0]=0.35
    for i in range(n):
        y[i+1] = a*y[i]*(1-y[i])
        x[i+1] = i
        w[i+1] = i/n
        v[i+1] = a*w[i]*(1-w[i])
    return v,w,x,y
```

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

v1,w1,x1,y1, = iterate(0.5)
v2,w2,x2,y2 = iterate(1.5)
v3,w3,x3,y3 = iterate(3.3)
v4,w4,x4,y4 = iterate(4.0)

plt.xlabel("n")
plt.ylabel("x[n]")
plt.plot(x1,y1,label="a=0.5")
plt.plot(x2,y2,label="a=1.5")
plt.plot(x3,y3,label="a=3.3")
plt.plot(x4,y4,label="a=4.0")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=2, mode="expand", borderaxespad=0.)
plt.savefig('sensitivity.png')
plt.show()

plt.xlabel("x[n]")
plt.ylabel("x[n+1]")
plt.plot(w1,v1,label="a=0.5")
plt.plot(w2,v2,label="a=1.5")
plt.plot(w3,v3,label="a=3.3")
plt.plot(w4,v4,label="a=4.0")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=2, mode="expand", borderaxespad=0.)
plt.savefig('logistic.png')
plt.show()
```
